**WEATHER FORECASTING ANALYSIS**

**IMPORTING LIBRARIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading the dataset**

```python
df = pd.read_csv("WEATHER_FORECASTING_ANALYSIS_DATA.csv")
print("Data set loaded successfully")
```

```python
print(df)
```

****

```python
# Query 1: Remove the first row (contains units like °C, %, mm etc.)
df = df.drop(index=0).reset_index(drop=True)
print(df)
```

```python
# Query 2: Convert numeric fields from string (object) to float
numeric_columns = [
    'Temperature', 'Humidity', 'Wind Speed', 'Wind Direction',
    'Precipitation', 'Atmospheric Pressure', 'Visibility',
    'Dew Point', 'Cloud Cover', 'UV Index', 'Solar Radiation',
    'Sea Level Pressure', 'Ground Temperature', 'Evapotranspiration',
    'Frost Point', 'Heat Index', 'Wind Chill', 'Air Quality Index'
]
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
print(df)
```

```python
# Query 3: Convert 'Date' column to datetime format
df["Date"] = pd.to_datetime(df["Date"], errors='coerce', dayfirst=True)
print(df)
```

```python
print("✅ Data Cleaning Complete")
print(df)
df.to_csv("cleaned_weather_data.csv", index=False)
```

**Data Filtering**

```python
# We’ll filter data where Precipitation > 50 mm, indicating heavy rainfall.
heavy_rain_df = df[df['Precipitation'] > 50]
print("Days with Heavy Rainfall (> 50 mm):", heavy_rain_df.shape[0])
print(heavy_rain_df[['Date', 'Location', 'Precipitation']].head())
```

```python
# We’ll filter rows where Temperature > 35°C and Humidity < 30%.
dry_heat_df = df[(df['Temperature'] > 35) & (df['Humidity'] < 30)]
print("Days with Dry Heat (Temp > 35°C and Humidity < 30%):", dry_heat_df.shape[0])
print(dry_heat_df[['Date', 'Location', 'Temperature', 'Humidity']].head())
```

```python
# Average Temperature by Location
avg_temp_by_location = df.groupby("Location")["Temperature"].mean().sort_values(ascending=False)
print("Average Temperature by Location:\n", avg_temp_by_location)
```

```python
# Total Precipitation by Weather Condition
total_precip_by_weather = df.groupby("Weather Condition")["Precipitation"].sum().sort_values(ascending=False)
print("Total Precipitation by Weather Condition:\n", total_precip_by_weather)
```

```python
# Top 5 Hottest Days (Descending Temperature)
hottest_days = df.sort_values(by="Temperature", ascending=False)
print("Top 5 Hottest Days:\n", hottest_days[['Date', 'Location', 'Temperature']].head(5))
```

```python
# Top 5 Windy Days (Descending Wind Speed)
windiest_days = df.sort_values(by="Wind Speed", ascending=False)
print("Top 5 Windiest Days:\n", windiest_days[['Date', 'Location', 'Wind Speed']].head(5))
```

```python
# Aggregation Query: Min, Max, Mean, Mode, Variance
weather_columns = ['Temperature', 'Humidity', 'Wind Speed', 'Precipitation']

aggregated_stats = {
    "Minimum": df[weather_columns].min(),
    "Maximum": df[weather_columns].max(),
    "Mean": df[weather_columns].mean(),
    "Mode": df[weather_columns].mode().iloc[0],
    "Variance": df[weather_columns].var()
}

for stat, values in aggregated_stats.items():
    print(f"\n{stat}:\n{values}")

```

```python
# Locations with Max UV Index > 10 (High Risk)
high_uv_locations = df[df['UV Index'] > 10]['Location'].unique()
print("Locations with UV Index > 10:\n", high_uv_locations)
```

```python
# Correlation Between Temp and Humidity
correlation = df[['Temperature', 'Humidity']].corr()
print("Correlation between Temperature and Humidity:\n", correlation)
```

```python
# Days with Tornado Occurrence
tornado_days = df[df['Tornado Occurrence'] == 'Yes']
print("🔹 Tornado Days Found:", tornado_days.shape[0])
```

**DATA VISUALIZATION**

```python
# 1. Line Plot of Temperature Over Time
# Shows how temperature changes daily/monthly.
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Date', y='Temperature', hue='Location')
plt.title('Temperature Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend(title='Location')
plt.show()
```

```python
# 2. Histogram of Humidity Distribution
# Visualizes the distribution of humidity values.
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Humidity', bins=30, kde=True)
plt.title('Humidity Distribution')
plt.xlabel('Humidity (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Boxplot of Wind Speed by Location
# Compares wind speed variability across different locations.
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Location', y='Wind Speed')
plt.title('Wind Speed Distribution by Location')
plt.xlabel('Location')
plt.ylabel('Wind Speed (km/h)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Heatmap of Correlation Between Weather Parameters
#Shows correlations among temperature, humidity, wind speed, precipitation, pressure, etc.
plt.figure(figsize=(8,6))
corr = df[['Temperature', 'Humidity', 'Wind Speed', 'Precipitation', 'Atmospheric Pressure']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Weather Parameters')
plt.show()
```

```python
# 5. Scatter Plot of Temperature vs. Humidity
# Explores the relationship between temperature and humidity.
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Temperature', y='Humidity', hue='Location', alpha=0.6)
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.legend(title='Location')
plt.show()
```
